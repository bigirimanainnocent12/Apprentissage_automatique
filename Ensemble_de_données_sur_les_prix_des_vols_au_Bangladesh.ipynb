# **Aperçu de l'ensemble de données : Ensemble de données sur les prix des vols au Bangladesh**

# Introduction

Le « Bangladesh Flight Fare Dataset » est un ensemble de données synthétiques comprenant 57 000Benregistrements de vols, conçus pour représenter des scénarios de transport aérien au départ du Bangladesh. Cet ensemble de données simule une dynamique réaliste des tarifs aériens, en prenant en compte des facteurs clés tels que les opérations aériennes, les spécificités des aéroports, les classes de voyage, les habitudes de réservation et les variations saisonnières propres au marché aérien du Bangladesh. Il est destiné aux chercheurs, data scientists et analystes intéressés par la prévision des tarifs aériens, l'analyse des habitudes de voyage ou les applications d'apprentissage automatique et d'apprentissage profond. En combinant des distributions statistiques inspirées du monde réel et les normes du secteur aéronautique, cet ensemble de données fournit une base solide pour l'étude de l'économie du transport aérien dans le contexte sud-asiatique.

# Objectif de l'ensemble de données

Cet ensemble de données vise à :

• Faciliter la modélisation prédictive des tarifs de vol, avec « Tarif total (BDT) » comme variable cible principale.

• Permettre l'analyse des tendances de voyage, y compris l'impact des festivals culturels (par exemple, l'Aïd, le Hajj) et les horaires de réservation sur les prix.

• Servir de ressource de formation pour les modèles d'apprentissage automatique (ML) et d'apprentissage profond (DL), avec une taille d'échantillon suffisante (50 000) et une diversité de fonctionnalités pour la généralisation.

• Fournir une représentation réaliste mais synthétique de l'écosystème du transport aérien au Bangladesh, combinant des scénarios de vol nationaux et internationaux.

# Collecte de données et méthodologie

L'ensemble de données est généré synthétiquement à l'aide de Python, sa méthodologie s'appuyant sur des données aéronautiques réelles et des principes statistiques. Vous trouverez ci-dessous une description détaillée de sa construction :

1. **Composants de données**

• Compagnies aériennes :

- Nombre : 25 compagnies aériennes (21 internationales, 4 nationales).

-Source : Compilé à partir de l'Autorité de l'aviation civile du Bangladesh et de l'historique des compagnies aériennes , y compris les principaux transporteurs comme Emirates, Qatar Airways et Biman Bangladesh Airlines.

- Sélection : Choix uniforme aléatoire par enregistrement de vol pour refléter la diversité opérationnelle.

• Aéroports :

- Aéroports sources : 8 aéroports nationaux (par exemple, DAC - aéroport
international Hazrat Shahjalal, Dhaka).

- Aéroports de destination : 20 aéroports (8 nationaux + 12 internationaux, par
exemple, DXB - Aéroport international de Dubaï).

- Coordonnées : Provenant des codes d'aéroport mondiaux , utilisés pour les calculs de distance.

- Noms complets : ajoutés pour plus de lisibilité, mappés via un dictionnaire (par exemple, « DAC » → « Aéroport international Hazrat Shahjalal, Dhaka »).

• Classes de voyage : Économie, Affaires, Première classe, standard dans l'ensemble du secteur, attribuées de manière aléatoire avec une distribution uniforme.

• Sources de réservation : réservation directe, agence de voyages, site Web en ligne, reflétant les méthodes courantes, selon Statista , avec une sélection aléatoire uniforme.

• Types d'avions : Boeing 777, Airbus A320, Boeing 737, Boeing 787, Airbus A350, attribués en fonction de la distance de vol, provenant de Boeing et Airbus .

2. Calculs clés

• Distance :

- Méthode : La formule Haversine calcule la distance du grand cercle :

a = sin²(Δφ/2) + cos(φ₁) cos(φ₂) sin²(Δλ/2)

c = 2 arctan2(√a, √(1-a))

d = R · c, R = 6371 km

• Objectif : Détermine la durée du vol, le type d'avion et les escales.

• Source : Wikipédia - Formule Haversine .

• Durée du vol :

-Formule : Durée = max(d/s · U(0,9, 1,1), 0,5), où s est la vitesse (300 km/h pour < 500 km, 600 km/h pour 500-2000 km, 900 km/h pour > 2000 km), et U est la variation aléatoire uniforme.

• Source : Vitesses ajustées à partir de World Atlas , garantissant le réalisme (par exemple, DAC vers CGP ~ 45 minutes).

• Tarifs :

• Tarifs de base :
• National : Économie (2000-5000 BDT), Affaires (5000-10000 BDT), Première classe (10000- 15000 BDT).

• International : Économie (5 000-70 000 BDT), Affaires (15 000-150 000 BDT), Première classe (25 000-300 000 BDT).

• Source : dérivé de Trip.com et Expedia , par exemple, DAC à LHR ~ 380-600 $ (~ 41 800-66 000 BDT à 1 USD = 110 BDT).

• Réglages :

• Multiplicateurs saisonniers (Régulier : 1,0, Aïd : 1,3, Hajj : 1,5, Hiver : 1,2), selon les tendances de la demande de Timeanddate.com .

• Jours avant le départ : 20 % de réduction (60 jours et plus), 10 % de réduction (30 à 59 jours), majoration de 20 % (< 5 jours), selon Skyscanner .

• Taxes : National : 200 BDT ; International : 2000-6000 BDT + 15 % du tarif de base,selon l'Autorité de l'aviation civile du Bangladesh .

• Formule : Tarif total = (Tarif de base · Multiplicateur · Remise/Majorité) + Taxe.

3. Méthodes statistiques

• Échantillonnage aléatoire : distributions uniformes utilisées pour la compagnie aérienne, la classe, la source de réservation et les jours avant le départ (1 à 90 jours), garantissant une représentation uniforme.

• Variabilité : La durée du vol inclut U(0,9, 1,1) pour modéliser les fluctuations du monde réel(par exemple, le vent, le routage).

• Escales : basées sur la probabilité (directes : 100 % pour < 1 000 km, 70 % pour 1 000 à 5 000 km ; 1 escale : 30 % pour 1 000 à 5 000 km, 80 % pour > 5 000 km ; 2 escales : 20 % pour > 5 000 km), reflétant les normes de l'industrie selon FlightConnections .

**Structure de l'ensemble de données**

• Lignes : 50 000 échantillons, suffisants pour la formation ML/DL et la robustesse statistique.

• Colonnes : 17 fonctionnalités :

1. Compagnie aérienne : Nom de la compagnie aérienne exploitante (par exemple, « Emirates
»).

2. Source : Code IATA de l'aéroport de départ (par exemple, « DAC »).

3. Nom de la source : Nom complet de l'aéroport de départ (par exemple, « Aéroport

5. Nom de la destination : Nom complet de l'aéroport d'arrivée (par exemple, « Aéroport de
Londres Heathrow »).
6. Date et heure de départ : Horodatage de départ (par exemple, « 2025-03-15 14:30:00 »).
7. Date et heure d'arrivée : Horodatage d'arrivée (par exemple, « 2025-03-16 02:45:00 »).
8. Durée (heures) : Durée du vol en heures (par exemple, 12,25).
9. Escales : Nombre d'arrêts (par exemple, « Direct », « 1 arrêt »).
10. Type d'avion : Modèle d'avion (par exemple, « Boeing 777 »).
11. Classe : Classe de voyage (par exemple, « Économique »).
12. Source de réservation : méthode de réservation (par exemple, « site Web en ligne »).
13. Tarif de base (BDT) : Prix de base du billet en taka bangladais (par exemple, 45 000).
14. Taxe et supplément (BDT) : Frais supplémentaires (par exemple, 6750).
15. Tarif total (BDT) : Prix final (par exemple, 51 750).
16. Saisonnalité : Saison des voyages (par exemple, « Aïd »).
17. Jours avant le départ : jours entre la réservation et le départ (par exemple, 45).

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for i, column in enumerate(['Airline', 'Aircraft Type', 'Class', 'Booking Source', 'Destination', 'Source', 'Seasonality']):
    sns.boxplot(x=column, y='Total Fare (BDT)', data=df, ax=axes[i])
    axes[i].set_title(f'Box plot de Total Fare (BDT) en fonction de {column}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
